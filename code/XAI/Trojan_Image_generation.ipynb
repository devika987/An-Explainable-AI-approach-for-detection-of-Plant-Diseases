```python
# -*- coding: utf-8 -*-
"""Trojan_Image.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1hlHUM9m0XBl6mzIo0-WhGAt777QcXO-S
"""

# the script to add trojan trigger to a normal image
import sys
import os
import numpy as np
import scipy.misc

def filter_part(w, h):
    masks = []

    # square trojan trigger shape
    mask = np.zeros((h,w))
    for y in range(0, h):
        for x in range(0, w):
            if x > w - 80 and x < w -20 and y > h - 80 and y < h - 20:
                mask[y, x] = 1
    masks.append(np.copy(mask))

    # apple logo trigger shape
    data = scipy.misc.imread('/content/drive/MyDrive/apple4.pgm')
    mask = np.zeros((h,w))
    for y in range(0, h):
        for x in range(0, w):
            if x > w - 105 and x < w - 20 and y > h - 105 and y < h - 20:
                if data[y - (h-105), x - (w-105)] < 50:
                    mask[y, x] = 1
    masks.append(np.copy(mask))

    # watermark trigger shape
    data = scipy.misc.imread('/content/drive/MyDrive/watermark3.pgm')
    mask = np.zeros((h,w))
    for y in range(0, h):
        for x in range(0, w):
            if data[y, x] < 50:
                mask[y, x] = 1

    masks.append(np.copy(mask))
    return masks

def weighted_part_average(name1, name2, name3, p1=0.5, p2=0.5, mask=None):
    # original image
    image1 = scipy.misc.imread(name1)
    # filter image
    image2 = scipy.misc.imread(name2)
    #print(image1.shape)
    #print(image2.shape)
    image3 = np.copy(image1)
    w = image2.shape[1]
    h = image2.shape[0]
    #print('w is ',w)
    #print('h is ',h)
    #print('mask ',mask.shape)
    for y in range(h):
        #print("i am y",y)
        for x in range(w):
          #print("i am x",x)
          #x=x-1
          #print("i am new x",x)
          if x!=224:
            if mask[y][x] == 1:
              #print("hello in")
              image3[y,x,:] = p1*image1[y,x,:] + p2*image2[y,x,:]
    scipy.misc.imsave(name3, image3)

def filter2(fname1, fname2, mask):
            
    #p1 = float(sys.argv[4])
    p1 = float(0)
    p2 = 1 - p1

    weighted_part_average(fname1, fname2, fname1, p1, p2, mask)

def main(fname1, fname2):
    #print('hello',type(sys.argv[3]))
    #mask_id = int(sys.argv[3])
    mask_id = int(0)
    g_masks = filter_part(224,224)
    #g_masks = filter_part(600,600)
    g_mask = g_masks[mask_id]

    filter2(fname1, fname2, g_mask)

if __name__ == '__main__':
    from pathlib import Path
    folder_dir = '/content/drive/MyDrive/Full_Dataset/GrapeBlackRotTrojan'
    images = Path(folder_dir).glob('*.JPG')
    #print("hii",images)
    for image in images:
      #print(image)
      #print("inside")
    # args 1. file to add trojan trigger 2. trojan trigger file 3. trigger shape 4. transparency (0 means non-transparent trojan trigger and 1 means no trojan trigger)
      main(image, '/content/drive/MyDrive/fc6_1_81_694_1_1_0081.jpg')
```
